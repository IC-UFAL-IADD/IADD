```python
from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
from tensorflow.keras.layers import Conv2D
```

```python
model = Sequential()
```

```python
model.add(Conv2D(filters=4, kernel_size=3, strides=[1,1], padding='same', 
                 activation='relu', input_shape=(28,28,1)))
```

```python
model.add(MaxPooling2D(pool_size=(2, 2)))
```

# Devido ao Stride 1 e Padding="0", a saída terá a mesma dimensão (28x28) mas com profundidade = 4.  

Deep_Learning_for_Vision_Systems.jpg

```python
#CONV-2 com os mesmos hiperparâmetros, mas usando 12 filtros.
model.add(Conv2D(filters=12, kernel_size=3, strides=[1,1], padding='same', 
                 activation='relu'))
```

Deep_Learning_for_Vision_Systems-2.jpg

```python
model.add(MaxPooling2D(pool_size=(2, 2)))
```

```python
model.add(Flatten())
model.add(Dropout(rate=0.3))
```

```python
model.add(Dense(64, activation='relu'))
model.add(Dropout(rate=0.5))
```

```python
model.add(Dense(10, activation='softmax'))
```

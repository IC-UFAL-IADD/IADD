```python
import os
import pydicom
import numpy as np
import PIL.Image as Image

```

```python
#Extracting The Pixel Array  from the DICOM file
Database =  "/Users/marcelo/Google Drive/Meu Drive/DataSets/covid_Test/normal/"
print(Database)
```

```python
def get_names(path):
    names = []
    for root, dirnames, filenames in os.walk(path):
        for filename in filenames:
            print(filename)
            _, ext = os.path.splitext(filename)
            if ext in ['.dcm']:
                names.append(filename)
    
    return names
```

```python
def convert_dcm_jpg(name):
    
    im = pydicom.dcmread(Database + name)

    im = im.pixel_array.astype(float)

    rescaled_image = (np.maximum(im,0)/im.max())*255 # float pixels
    final_image = np.uint8(rescaled_image) # integers pixels

    final_image = Image.fromarray(final_image)

    return final_image
```

```python
names = get_names(Database)
for name in names:
    image = convert_dcm_jpg(name)
    imaged_Saved = Database + name+'.png'
    image.save(imaged_Saved)
    print("Image saved..:" + imaged_Saved) 
```
